```python
import os
import pandas as pd
```

```python

df_w=pd.read_excel('pulse_prod_config_audit_2022-07-15.xlsx', sheet_name='terminal_prov_audit_rpt')
print(df_w)
```

```python
df=df_w[['terminal_name', 'obj_id', 'serialnumber']]
df.reset_index(drop=True)
print(df)
```

```python
terminals_list=df.to_dict('records')
print(terminals_list)
```

```python
terminal_obj_id_dict={}
for i in terminals_list:
    
    terminal=i.get('terminal_name')
    obj_id=i.get('obj_id')
    serial=i.get('serialnumber')
    terminal_obj_id_dict[int(serial)]=[terminal,obj_id]
print(terminal_obj_id_dict)
        
    
```

```python
import json
with open("terminal_objid_map.json", "w") as outfile:
    json.dump(terminal_obj_id_dict, outfile)
```
